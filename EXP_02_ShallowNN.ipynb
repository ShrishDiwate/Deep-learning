```python
import tensorflow as tf

# 1. Load and normalize the data in fewer lines
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# 2. Define, compile, and train the model
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# The fit() method trains the model. We can hide the per-epoch output with verbose=0.
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test), verbose=0)

# 3. Evaluate the model and print the final accuracy
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Final Test Accuracy: {accuracy * 100:.2f}%")
```
