```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

a = tf.constant(1.8)
b = tf.constant(3.6)

node_round = tf.round(a)
node_floor = tf.floor(b)

node_add = tf.add(node_round, node_floor)
node_abs = tf.abs(node_add)

node_add = tf.add(node_round, node_floor)
node_abs = tf.abs(node_add)

node_x = tf.multiply(node_round, 5.0)

final_result_node = tf.add(node_x, node_abs)
with tf.Session() as sess:
    output = sess.run(final_result_node)
    print(f"The final output of the graph is: {output}")
```
